# 변수와 상수

```python
import numpy as np
import matplotlib.pyplot as plt

a=1.5
x=np.linspace(-1,1)
y=a*x

plt.plot(x,y)
plt.xlabel("x",size=14)
plt.ylabel("y",size=14)
plt.grid()
plt.show()
```

# 지수함수

```python
def my_func(x):
    a=3
    return x**a

x=np.linspace(0,2)
y=my_func(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

# 무리함수

```python
def my_func(x):
    return np.sqrt(x)+1
x=np.linspace(0,2)
y=my_func(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

# 다항함수"

```python
def my_func(x):
    return 3*x**2+2*x+1
x=np.linspace(-1,1)
y=my_func(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

```python
def my_func(x):
    return 4*x**3+2*x**2+x+3

x=np.linspace(-1,1)
y=my_func(x)

plt.plot(x,y)
plt.show()
```

```python
def my_func(x):
    return x**3-2*x**2-3*x+4

x=np.linspace(-1,1)
y=my_func(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

# 삼각함수

```python
def my_func(x):
    return np.sin(x)
def my_cos(x):
    return np.cos(x)
x=np.linspace(-np.pi, np.pi)
y_sin=my_func(x)
y_cos=my_cos(x)

plt.plot(x,y_sin,label="sin")
plt.plot(x,y_cos,label="cos")
plt.legend()
plt.grid()
plt.show()
```

```python
def my_tan(x):
    return np.tan(x)

x=np.linspace(-1.3,1.3)
y=my_tan(x)

plt.plot(x,y,label="tan")
plt.legend()
plt.grid()
plt.show()
```

# sin cos 함수의 절댓값

```python
x=np.linspace(-np.pi, np.pi)
y_sin=np.abs(np.sin(x))
y_cos=np.abs(np.cos(x))

plt.scatter(x,y_sin, label="Sin")
plt.scatter(x,y_cos, label="Cos")
plt.legend()
plt.grid()
plt.show()
```
